# **Credit card Fraud detection**

```python
#Import dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score,confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load the dataset
df=pd.read_csv("creditcard.csv")
```

```python
#Dataset is highly unbalanced
df["Class"].value_counts()
#Seperating for analysis
legit=df[df.Class==0]
fraud=df[df.Class==1]
```

# **Under-Sampling**

Undersampling is a technique used to handle imbalanced datasets, where one class (usually the negative class, like "not fraud") has many more samples than the other class (usually the positive class, like "fraud").

**Why Use Undersampling?**

In imbalanced datasets:

The model might get biased toward the majority class.

It might achieve high accuracy by just predicting the majority class but perform poorly on the minority class, which is usually more important.


**How Undersampling Works**

It reduces the number of samples in the majority class to match (or come closer to) the number of samples in the minority class.

Example:

If your dataset has:

10,000 instances of class 0

500 instances of class 1

Undersampling might randomly reduce class 0 to 500 instances so the model trains on:

500 class 0

500 class 1

```python
#Build a dataset containing similar no of legit and fraudulent
legit_sample=legit.sample(n=492)
print(legit_sample.shape)
```

```python
#Concatenating Dataframe
new_df=pd.concat([legit_sample,fraud],axis=0)
new_df.head()
```

**Vertical and horizantal split of data**

```python
#Seperate features and labels
X=new_df.drop(columns='Class',axis=1)
y=new_df['Class']
#Train test split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=2)
```

# **Model Training**

```python
#Model--->Logistic Regression Model
model=LogisticRegression()
```

```python
#Fit Model
model.fit(X_train,y_train)
```

```python
#Predict based on train data
x_train_predict=model.predict(X_train)
```

**Model accuracy based on Train-data Accuracy**

```python
x_train_accuracy=accuracy_score(x_train_predict,y_train)
print("Accuracy on train data out of 1:",x_train_accuracy)
```

**Model accuracy based on Test-data Accuracy**

```python
#Predict based on test data
x_test_predict=model.predict(X_test)
x_test_accuracy=accuracy_score(x_test_predict,y_test)
print("Accuracy on test data out of 1:",x_test_accuracy)
```

# Classification Report

```python
#Classification Report....
print("Classification Report.....")
#Finding Accuracy
x_test_accuracy=accuracy_score(x_test_predict,y_test)
print("Accuracy on test data out of 1:",x_test_accuracy)
#Finding Precision
precision=precision_score(x_test_predict,y_test)
#Finding Recall
recall=recall_score(x_test_predict,y_test)
#Finding F1 score
f1=f1_score(x_test_predict,y_test)
#Printing the results
print("Precision Score:",precision)
print("Recall Score:",recall)
print("F1 Score:",f1)
#Ploting Confusion Matrix
cm=confusion_matrix(x_test_predict,y_test)
sns.heatmap(cm,annot=True)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

**Explanation**

Accuracy Score: 0.9441

This means that 94.41% of the total predictions were correct. It gives an overall idea of how often the classifier is correct.

Precision Score: 0.9183

This indicates that 91.83% of the instances predicted as positive (class 1) were actually positive. A high precision score means fewer false positives.

Recall Score: 0.9677

Out of all actual positive instances, 96.77% were correctly identified by the model. This high recall is particularly valuable when missing a positive case is costly (e.g., fraud detection, disease diagnosis).

F1 Score: 0.9424

The F1 Score is the harmonic mean of precision and recall, giving a single metric that balances both. An F1 score of 94.24% shows that your model performs well in scenarios where both false positives and false negatives are important.
